```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
data = pd.read_csv('/home/ds/Downloads/insurance.csv', index_col = 0)
```

```python
data.head()
```

```python
data.columns = ['quarter', 'eid', 'sex', 'race', 'age', 'visited', 'salary', 'score']
```

```python
fig, axs = plt.subplots(1, 5, sharey=True)
data.plot(kind='scatter', x='quarter', y='score', ax=axs[0], figsize=(16, 8))
data.plot(kind='scatter', x='age', y='score', ax=axs[1])
data.plot(kind='scatter', x='salary', y='score', ax=axs[2])
data.plot(kind='scatter', x='visited', y='score', ax=axs[3])
data.plot(kind='scatter', x='sex', y='score', ax=axs[4])
```

```python
import statsmodels.formula.api as smf

# create a fitted model in one line
lm = smf.ols(formula='score ~ age', data=data).fit()

# print the coefficients
lm.params
lm.summary()
```

```python
lm = smf.ols(formula='score ~ salary', data=data).fit()

# print the coefficients
lm.params
lm.summary()
```

```python
lm = smf.ols(formula='score ~ visited', data=data).fit()

# print the coefficients
lm.params
lm.summary()
```

```python
lm = smf.ols(formula='score ~ visited+sex+race', data=data).fit()

# print the coefficients
lm.params
```

```python
lm.summary()
lm.rsquared
```

```python
lm = smf.ols(formula='score ~ age+salary+sex+race+visited', data=data).fit()

# print the coefficients
lm.params
lm.pvalues
```

```python
lm.summary()
```

```python
lm.rsquared
```

```python
lm = smf.ols(formula='score ~ age+sex+race+visited', data=data).fit()

# print the coefficients
lm.summary()
lm.pvalues
```

```python
lm = smf.ols(formula='score ~ age+sex+visited', data=data).fit()

# print the coefficients
lm.params
lm.pvalues
```

```python
lm.summary()
```

```python
lm = smf.ols(formula='score ~ quarter+visited+age+sex', data=data).fit()

# print the coefficients
lm.summary()
lm.pvalues
```
